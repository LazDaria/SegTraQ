```python
import matplotlib.pyplot as plt
import spatialdata_plot

fig, axes = plt.subplots(1, 1, figsize=(5, 5))

sdata_r1.pl.render_shapes(
    #color='centroid_x',
    contour_px=1,
    fill_alpha=0.3,
).pl.show(ax=axes)

axes.set_title('Cell size on shapes')

plt.tight_layout()
```

```python
from spatialdata import bounding_box_query

def crop0(x):
    return bounding_box_query(
        x,
        min_coordinate=[1520, 2750],
        max_coordinate=[1711, 2912],
        axes=("x", "y"),
        target_coordinate_system="global",
    )


sdata_subset = crop0(sdata_r1)
```

```python
sdata_subset
```

```python
import matplotlib.pyplot as plt
import spatialdata_plot

fig, axes = plt.subplots(1, 1, figsize=(5, 5))

sdata_subset.pl.render_labels(
    color='centroid_x',
    contour_px=1,
    fill_alpha=0.3,
).pl.show(ax=axes)

axes.set_title('Cell size on labels')

plt.tight_layout()
```

```python
gene_name = 'EPCAM'

sdata_subset.tables["table"].obs["region"] = "cell_boundaries"
sdata_subset.set_table_annotates_spatialelement("table", region="cell_boundaries")

sdata_subset.pl.render_images("morphology_focus").pl.render_shapes(
    "cell_boundaries",
    color=gene_name,
).pl.render_points(
    "transcripts",
    color="feature_name",
    groups=gene_name,
    palette="orange",
).pl.show(title=f"{gene_name} expression over H&E image", coordinate_systems="global", figsize=(10, 5))
```

```python
import matplotlib.pyplot as plt
import spatialdata_plot

sdata_subset.tables["table"].obs["region"] = "cell_labels"
sdata_subset.set_table_annotates_spatialelement("table", region="cell_labels")

fig, axes = plt.subplots(1, 1, figsize=(5, 5))

sdata_subset.pl.render_images("morphology_focus").pl.render_labels(
    "cell_labels",
    color=gene_name,
).pl.render_points(
    "transcripts",
    color="feature_name",
    groups=gene_name,
    palette="orange",
).pl.show(title=f"{gene_name} expression over H&E image", coordinate_systems="global", figsize=(10, 5))
```
